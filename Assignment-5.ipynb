```python
#Import Libraries
import glob
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
```

### All required functions are in the cells below

```python
#Function to check if there are any problems in the dataset    
def identify_issues(datasets):
    df = pd.read_csv(datasets)
    if df[df.keys()[0]].max() == 0 and df[df.keys()[20]].max() == 20:
        print('I am uncertain about the dataset \n')
    else:
        mins = np.array([df[key].min() for key in df.keys()])
        minssum = mins.sum()
        if minssum == 0:
            print('It is an error \n')
        else:
            print('Dataset is fine \n')
#Function for visualize the dataset with respect to min max and average
def visualize(datasets):
    df = pd.read_csv(datasets)
    template_df = df
    template_dict = dict()
    # Note down Average, Maximum, Minimum for each feature in a dataset
    for key in template_df.keys():
        template_dict[key] = list()
        template_dict[key].append(template_df[key].mean())
        template_dict[key].append(template_df[key].max())
        template_dict[key].append(template_df[key].min())
    finalized_df = pd.DataFrame.from_dict(template_dict).T
    finalized_df.columns = ['Average','max','min']
    #Creating subplots for visualization
    t,ax = plt.subplots(1,3,figsize = [15,15])
    s = 0
    for col in finalized_df.keys():
        ax[s].plot(np.array(finalized_df.index),np.array(finalized_df[col]))
        ax[s].set_title(col)
        s = s + 1 
    plt.show(t)
```

### Run the code Below

```python
#Load Datasets
dataset = sorted(glob.glob('data*.csv'))
for datasets in dataset[:4]:
    print(datasets)
    visualize(datasets)
    identify_issues(datasets)
```
